**Limpeza de Dados**

```python
# Procurar valores vazios
# criar copia do df
data_pre = df.copy()
data_pre
```

```python
# conversao de dados

# Em uma dataframe, se na coluna ele ver q tem varios valores numericos, mas ver q tem 1 string ele muda tudo para object

# Converter data, para isso o dado da data, deve ser no modelo YYYY-MM-DD ex: 2020-01-15
data_pre['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'], errors='coerce')
data_pre['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'], errors='coerce')

# Aqui modificamos o valor da coluna com datas, para datetime
data_pre.loc[df['DATA FINAL'] > '2010-01-01']

# Quando se converte para datetime, o pandas deixa fazer esse tipo de comparacao, pois ele considera aquela string como uma data mesmo

data_pre.info()
```

```python
# Conversao para dados númericos
lista = ['MARGEM MÉDIA REVENDA','PREÇO MÉDIO DISTRIBUIÇÃO', 'DESVIO PADRÃO DISTRIBUIÇÃO', 'PREÇO MÍNIMO DISTRIBUIÇÃO','PREÇO MÁXIMO DISTRIBUIÇÃO']

for i in lista:
    data_pre[i] = pd.to_numeric(df[i], errors='coerce')

# Aqui eu converto o tipo de dados para numerico, onde oq ele faz é verificar valores string q pode ser convertido para numerico e os q nao, ele explode um erro q vc define
# Erros pode ser 
# Coerce - Onde n foi possível converter ele retorna 'NaN'
# Raise - Onde ele explode e para de rodar e volta um erro
data_pre.info()
```

```python
# Verificar os valores q ele considera nulos, onde for ele retorna NaN
# Aqui ele retorna uma lista de booleanos com true e false, mostrando o index q está com o valor nulo
lista_booleanos_valores_nulos = data_pre['MARGEM MÉDIA REVENDA'].isnull()
data_pre[lista_booleanos_valores_nulos]
```

```python
# Preencher os valores NaN com um valor padrão, onde tudo nan será esse tal valor
data_pre_fill = data_pre.fillna(0) # Aqui oq ele faz é o seguinte ele vai passar para todos os valores nan, o valor q vc passar para ele
# ele retorna uma CÓPIA da df, ou seja nao muda o df original, caso for no df original precis por inplace=true

data_pre_fill[lista_booleanos_valores_nulos] # aqui eu so coloco o filtro para pode ver o resultado
```

**Iterar sobre o dataframe com o pandas sem iterrows**

```python
df = pd.DataFrame({
    'A':[1,2,3,4],
    'B': [10,20,30,40],
    'C': [100,200,300,400]
},index=['linha 1', 'linha 2', 'linha 3', 'linha 4'])
```

```python
# Método apply ele aplica uma função para cada linha inteira, ou coluna inteira ou a cada elemento do Dataframe
# 2 eixos
# Onde o axis=0 ele vai pegar e aplicar a função na coluna toda
# axis=1 ele vai pegar e aplicar a função na linha toda
```

```python
# Fazer a soma de A, B e C por linha
# Cria a função q desejo aplicar na linha
def soma(dado):
    print(dado) # aqui é meio q um dicionario, onde o dado seria o dic, e o ['A'] seria o index/chave de qual valor q é
    print(dado['A'])
    # Aqui ele pega o dado, no caso a linha toda, e o A mostra o valor so da coluna A, ele recebe a linha toda, onde eu filtro ali so para tal coluna
    # return dado['A'] + dado['B'] + dado['C']

df['Valor soma'] = df.apply(soma, axis=1)
df
```

```python
# Usar o axis=0 ele percorre toda as colunas e faz a função
valor2 = df[['A','B','C']].apply(soma, axis=0) # So pego as colunas q eu quero, pq senão ele pega tudinho

# colocar numa linha nova
df.loc['linha 5'] = valor2 # aqui uso o loc para ele pegar a linha do index linha 5, como n existe ele cria e ja poe os dados
df
```

```python
# Função para tirar a média tanto por linha e colua
def media(dado):
    return dado.mean()

valor3 = df[['A','B','C']].apply(media, axis=1)
df['Média'] = valor3
df
```

```python
# Nesse caso vem meio q um dic, onde coluna é um dic mas agora a chave e o index da linha ja q o dic agr é coluna interia
def teste(coluna):
    print(coluna)
    print(coluna['linha 1']) # Aqui retorna o index 0 de cada coluna ou ent o nome do index do dataframe, ai ele filtra, na coluna tal a lihha tal

df.apply(teste, axis=0)
```
