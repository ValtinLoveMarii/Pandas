```python
# EXE1 - Use applymap() para converter todos os números em strings formatadas como moeda (ex: "R$ 12.000,00").
df = pd.DataFrame({
    'vendas_2024': [12000, 35000, 8700],
    'vendas_2025': [15000, 42000, 9100]
})

def converter(valor):
    float(valor)
    valor_formatado = f'R$ {valor:,.2f}'
    new = valor_formatado.replace(',', '.')
    return new
    
    
df.applymap(converter)
```

```python
# EXE2 - Use map() para contar o número de palavras em cada frase. Depois, use outra função para retornar o texto em MAIÚSCULAS se tiver mais de 3 palavras.
textos = pd.Series([
    'Python é incrível',
    'pandas é ótimo para dados',
    'amo aprender coisas novas'
])
def maiuscula(valor):
    novo_valor = valor.split()
    new = len(novo_valor)
    if new > 3:
        return valor.upper()

textos.map(maiuscula)
```

```python
# EXE3 - Crie uma função e use applymap() para: Substituir None por "Desconhecido" se for texto, Substituir None por 0 se for número.
df = pd.DataFrame({
    'nome': ['Ana', 'Beto', None],
    'idade': [23, None, 35],
    'cidade': ['SP', 'RJ', 'BH']
})

def nulo(valor):
    if pd.isna(valor):
        if valor == None:
           return 'desconhecido'
        elif isinstance(valor, float):
            return 0
        else:
            print('errado')
    else:
        return valor
df.applymap(nulo)
```

```python
# EXE4 - Mostre apenas o maior salário de cada departamento (.max()). Agrupe por Departamento e retorne o menor e maior salário usando .agg(['min', 'max']). Use o atributo .groups pra visualizar os índices das linhas pertencentes a cada grupo.
df = pd.DataFrame({
    'Departamento':['RH', 'RH', 'TI', 'TI', 'Vendas', 'Vendas'],
    'Salário': [2000, 2500, 3000, 4000, 1800, 2200]
})

grupo_departamento = df.groupby('Departamento').sum()
media = grupo_departamento.mean()
conta = df.groupby('Departamento')
conta.count()
```

```python
# EXE 15 - Crie um DataFrame com strings e números misturados. Use applymap() para deixar em maiúsculas apenas as células do tipo str.
num_text = pd.DataFrame({
    'col1': [1, 'num', 3],
    'col2': [4, 5, 'mum']
})

def teste(valor):
    if type(valor) is int:
        return valor
    else:
        return valor.upper()
num_text.applymap(teste)
```

```python
# EXE 16 - Crie um DataFrame com valores numéricos e aplique applymap() para: Substituir números pares por "PAR" e ímpares por "ÍMPAR".
numeros = pd.DataFrame({
    'n1': [1, 2, 3, 4, 5],
    'n2': [6, 7, 8, 9, 10]
})

def teste2(valor):
    if valor%2 == 0:
        return 'PAR'
    else:
        return 'ÍMPAR'
    
numeros.applymap(teste2)
```

```python
# EXE 17 - Combine applymap() com uma função personalizada: Crie uma função que retorna "OK" se o valor for positivo e "NEG" se for negativo.Aplique-a a todo o DataFrame.
def func_generica(valor):
    if valor < 0:
        return 'NEG'
    else:
        return 'OK'

numeros.applymap(func_generica)
```
