```python
import pandas as pd
import numpy as np
```

```python
# EXE1 - Use applymap() para converter todos os números em strings formatadas como moeda (ex: "R$ 12.000,00").
df = pd.DataFrame({
    'vendas_2024': [12000, 35000, 8700],
    'vendas_2025': [15000, 42000, 9100]
})

def converter(valor):
    float(valor)
    valor_formatado = f'R$ {valor:,.2f}'
    new = valor_formatado.replace(',', '.')
    return new
    
    
df.applymap(converter)
```

```python
# EXE2 - Use map() para contar o número de palavras em cada frase. Depois, use outra função para retornar o texto em MAIÚSCULAS se tiver mais de 3 palavras.
textos = pd.Series([
    'Python é incrível',
    'pandas é ótimo para dados',
    'amo aprender coisas novas'
])
def maiuscula(valor):
    novo_valor = valor.split()
    new = len(novo_valor)
    if new > 3:
        return valor.upper()

textos.map(maiuscula)
```

```python
# EXE3 - Crie uma função e use applymap() para: Substituir None por "Desconhecido" se for texto, Substituir None por 0 se for número.
df = pd.DataFrame({
    'nome': ['Ana', 'Beto', None],
    'idade': [23, None, 35],
    'cidade': ['SP', 'RJ', 'BH']
})

def nulo(valor):
    if pd.isna(valor):
        if valor == None:
           return 'desconhecido'
        elif isinstance(valor, float):
            return 0
        else:
            print('errado')
    else:
        return valor
df.applymap(nulo)
```
