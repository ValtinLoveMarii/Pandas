```python
# EXE6 - Dada uma coluna "temp_C" (em Celsius), use .apply() para criar uma nova coluna "temp_F" (em Fahrenheit).
df_temp = pd.DataFrame({
    'Temp': [32, 12, 9, -1, 17]
})

def temp_f(valor):
    valor_f = (valor*1.8) + 32
    return valor_f
df_temp.apply(temp_f, axis=1)
```

```python
# EXE7 - Crie um DataFrame com colunas "nome" e "idade". Use .apply() para criar uma nova coluna "faixa_etaria" com três categorias: menor de idade, adulto, idoso.
df_nomes_idade = pd.DataFrame({
    'nome': ['Nome1', 'Nome2'],
    'idade': [19, 13]
})

def faixa_etaria(valor):
    if valor['idade'] < 18:
        return 'Menor'
    elif valor['idade'] >= 18 and valor['idade'] <= 50:
        return 'Maior'
    else:
        return 'Idoso'
    
a = df_nomes_idade.apply(faixa_etaria, axis=1)
df_nomes_idade['faixa_etaria'] = a

df_nomes_idade
```

```python
# EXE8 - Crie um DataFrame com colunas "nota1" e "nota2". Use .apply(axis=1) para calcular a média entre as duas e criar uma nova coluna "média".
df_notas = pd.DataFrame({
    'n1': [9, 7],
    'n2': [5, 10]
})

def media(valor):
    return valor.mean()

df_notas['media']=df_notas.apply(media, axis=1)
```

```python
# EXE9 - Crie colunas nota1, nota2 e nota3. Use apply() para calcular a média ponderada e classificar como 'aprovado' ou 'reprovado'.
df_notas_ponderada = pd.DataFrame({
    'n1': [9, 7],
    'n2': [5, 10],
    'n3': [4, 10]
})

def media_ponderada(valor):
    valor_media_ponderada =  (valor['n1'] + valor['n2'] + (valor['n3']*2)) / 4
    
    if valor_media_ponderada >= 6:
        return 'Aprovado'
    else:
        return 'Reprovado'


df_notas_ponderada['media'] = df_notas_ponderada.apply(media_ponderada, axis=1)
df_notas_ponderada
```

```python
# EXE10 - Crie uma coluna de salários e use apply() com lambda para aplicar um aumento de 10% apenas se o salário for menor que 3000.
df_salario = pd.DataFrame({
    'salario': [1500, 2000, 6700, 1230],
})

def aumento_salario(valor):
    if valor['salario'] <= 3000:
        porcentagem = (valor['salario'] / 100) * 10
        return valor + porcentagem
    else:
        return valor

df_salario.apply(aumento_salario, axis=1)
```

```python
# EXE11 - Crie uma função que receba uma linha (axis=1) e retorne duas informações novas (ex: media e desvio).
df_salario2 = pd.DataFrame({
    'salario': [1500, 2000, 6700, 1230],
    'salario2': [1430, 2340, 1700, 4230],
})

def media_desvio(valor):
    return valor.mean(), valor.std()
    
valores_brutos = df_salario2.apply(media_desvio, axis=1)
dic_values = valores_brutos.to_dict()
dic_certo = {}
for i, x in dic_values.items():
    dic_certo.update({i: [x[0], x[1]]})

lista_values_m = []
lista_values_d = []

for i, x in dic_certo.items():
    lista_values_m.append(x[0])
    lista_values_d.append(x[1])
    
    
df_salario2['media'] = lista_values_m

df_salario2['desvio'] = lista_values_d

df_salario2
```

```python
# EXE12 - Crie uma coluna de datas e use apply() para extrair o dia da semana e mês.
df_datas = pd.DataFrame({
    'data1': ['2020-01-01', '2022-02-02'],
    
})

df_datas['data1'] = pd.to_datetime(df_datas['data1'], errors='coerce')
# df_datas.info()

def datas_valores(valor):
   nome_semana = valor['data1'].day_name()
   nome_mes = valor['data1'].month_name()
   return nome_semana, nome_mes

valor_datas = df_datas.apply(datas_valores, axis=1)
valor_datas.to_dict()

lista_dias = []
lista_mes = []
for i, x in valor_datas.items():
    lista_dias.append(valor_datas[i][0])
    lista_mes.append(valor_datas[i][1])

print(lista_dias, lista_mes)

df_datas['dia_semana'] = lista_dias
df_datas['mes'] = lista_mes

df_datas
```

```python
# EXE19 - Use map() para classificar:
salarios = pd.Series([2500, 4000, 8000, 12000])
def func3(valor):
    if valor <= 3000:
        return 'Baixo'
    elif valor > 3000 and valor >= 8000:
        return 'Médio'
    else:
        return 'Alto'

salarios.map(func3)
```

```python
# EXE20 - Crie uma nova coluna total = preço × quantidade Depois, use map() ou apply() em df['total'] para aplicar 10% de desconto.
df = pd.DataFrame({
    'produto': ['A', 'B', 'C'],
    'preco': [10, 20, 30],
    'quantidade': [2, 5, 3]
})

df['total'] = df['preco'] * df['quantidade']

def func4(valor):
    per = (valor / 100) * 10
    return valor + per

df['total'] = df['total'].map(func4)
```
