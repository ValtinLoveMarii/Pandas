```python
# Criar o dataframe
import pandas as pd
import numpy as np
data = pd.DataFrame({
    'Nome': ['Naruto', 'Momo','Okarun', 'Jiji', 'Aira'],
    'Idade': [23, 16, 16, 17, ''],
    'Cidade': ['Konoha', 'Kamikoshi','Kamikoshi','Kamikoshi','Kamikoshi'],
    'Salario': [90000, 3400, 3400, 3000, 5000]
})

# Criando uma cópia para o tratamento de dados
df = data.copy()
df
```

```python
# Selecione múltiplas condições:Pessoas com idade > 30 e salário > 5000.
df.loc[(df['Idade'] > 18) & (df['Salario'] > 3000)]

```

```python
# Filtrar clientes cujo nome começa com a letra "A".
lista_nomes = []
for i, r in df.iterrows():
    if r['Nome'].startswith('A'):
        lista_nomes.append(True)
    else:
        lista_nomes.append(False)
    

print(lista_nomes)
df[lista_nomes]
```

```python
# Filtre apenas as pessoas que não moram em "Rio de Janeiro".
df.loc[~(df['Cidade'] == 'Kamikoshi')]
```
