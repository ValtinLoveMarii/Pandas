```python
# DataFrame novo
data = {
    "id": ["001", "002", "003", "004", "005"],
    "idade": ["25", "30", "vinte e cinco", "40 anos", None],
    "salario": ["3000.50", "4000", "cinco mil", "4500,75", "3800.20"],
    "data_contratacao": ["2021-05-12", "12/06/2020", "Junho 5, 2019", "2020-13-01", "2022/03/15"],
    "bonus": ["10%", "5%", "dez%", "15.5%", None],
}
data = pd.DataFrame(data)

data.info()
```

```python
# Converta a coluna id para o tipo inteiro.
data['id'] = pd.to_numeric(data['id'], errors='coerce')
```

```python
# Converta a coluna idade para numérico, transformando os valores inválidos em NaN.
data['idade'] = pd.to_numeric(data['idade'], errors='coerce')
data
```

```python
# Converta a coluna salario para float, substituindo vírgulas por pontos.
lista_com_novos_valores = []
for i, r in data.iterrows():
    valor = r['salario']
    
    if ',' in valor:
        valor = valor.replace(',','.')
        lista_com_novos_valores.append(valor)
    elif pd.isna(valor):
        print(f'Valor nulo {valor}')
        lista_com_novos_valores.append(np.nan)
    else:
        try:
            
            valor_numero_float = float(valor)
            print(f'VALOR PADRÃO VEIO ASSIM {valor}')
            lista_com_novos_valores.append(valor_numero_float)
        except:
            print(f'Valor númerico errado, ira ser NaN, na proxima direito! {valor}')
            lista_com_novos_valores.append(np.nan)

data['salario'] = lista_com_novos_valores
data.info()
    
```

```python
data
```

```python
# Converta a coluna data_contratacao para o tipo datetime, tratando os formatos diferentes.
formatos = [
    "%d/%m/%Y",
    "%Y-%m-%d",
    "%Y/%m/%d",
    "%m/%d/%Y",
    "%Y-%d-%m",
    "%Y/%d/%m",
    "%B %d, %Y",
    "%A, %d de %B de %Y",
    "%Y-%m-%d %H:%M",
    "%d/%m/%Y %H:%M:%S",
    "%I:%M %p",
    "%H:%M:%S"
]

lista_data_formata = []
for i, r in data.iterrows():
    valor_data_desformatada = r['data_contratacao']
    print(f'VALOR LÀ DE CIMA!!! {valor_data_desformatada}')
    valor_formatado = 0
    #verificar se é nulo
    if pd.isna(valor_data_desformatada):
        print(f'Valor nulo')
        lista_data_formata.append(valor_data_desformatada)
    
    for x in formatos:
        try:
            valor_formatado = datetime.strptime(valor_data_desformatada, x)
            # print(f'Consegui formatar {valor_formatado}')
            break
        except:
            continue
    if valor_formatado != 0:
        lista_data_formata.append(valor_formatado)
    else:
        lista_data_formata.append(np.nan)

data['data_contratacao'] = lista_data_formata
data
```

```python
# Calcule uma nova coluna chamada ano_contratacao com o ano de cada data.
lista_anos = []
for i, r in data.iterrows():
    valor_ano = r['data_contratacao']
    if pd.isna(valor_ano):
        print('VALOR NULO')
        lista_anos.append(np.nan)
    else:
        valor_ano = str(valor_ano)
        valor_ano = datetime.strptime(valor_ano, '%Y-%m-%d %H:%M:%S')
        valor_ano = valor_ano.year
        lista_anos.append(valor_ano)
    
```

```python
data
```

```python
# Converta a coluna bonus para float, removendo o símbolo % e dividindo por 100.
lista_bonus = []
for i, r in data.iterrows():
    valor_bonus = r['bonus']
    if pd.isna(valor_bonus):
        print('Valor nulo')
        lista_bonus.append(np.nan)
    else:
        if '%' in valor_bonus:
            valor_bonus = valor_bonus.replace('%', '')
            try:
                valor_bonus = pd.to_numeric(valor_bonus, errors='coerce', downcast='float')
                valor_final = valor_bonus / 100
                lista_bonus.append(valor_final)
            except:
                continue
        else:
            continue
            
data['bonus'] = lista_bonus
data
```

```python
# Substitua valores ausentes (NaN) na coluna idade pela média das idades válidas.
data.loc[data['idade'].isnull(), 'idade'] = data['idade'].mean()
data
```
